```python
import pandas as pd 
import plotly.graph_objs as go 
import datetime as dt
```

```python
data=pd.read_csv('stock(data).csv')
print(data)
```

```python
print(data.columns)
```

```python
print(data.head())
```

```python
data.columns = data.columns.str.strip()
```

```python
candlestick = go.Candlestick(x=data['Date'],
                             open=data['Open'],
                             high=data['High'],
                             low=data['Low'],
                             close=data['Close'])


layout = go.Layout(title='Candlestick Chart',
                   xaxis_title='Date',
                   yaxis_title='Price')


fig = go.Figure(data=[candlestick], layout=layout)


fig.show()
```

```python
fig.write_html('candlestick_chart.html')
```
